```python
driver = webdriver.Chrome(executable_path=binary_path)
actions.login(driver, email, password)

start_time = time.time()

for company in company_list:
    co = str(company).lower().strip()
    co = co.replace(" ", "-")
    for symbol in ["'", ',', '.']:
        co = co.replace(symbol, '')
    company_industry, company_size = get_company_info(co)
    connections_df.loc[connections_df['Company'] == company, 'Company Industry'] = company_industry
    connections_df.loc[connections_df['Company'] == company, 'Company Size'] = company_size
    
end_time = time.time()
runtime = end_time -start_time
runtime # 8018.039364099503
# This took about 2 hours and 23 mins to run
# Only run again on new raw df
```
